```python
import sqlite3
import pandas as pd

# Connect to SQLite database (it will create a new database file)
conn = sqlite3.connect("students.db")
cursor = conn.cursor()
```

```python
# SQL query to fetch students who scored above 80% with Grade 'A'
query_1 = '''
SELECT sm.StudentName, sm.Gender, sm.DOB, sm.FatherName, sm.MotherName, 
       sm.FathersOccupation, sm.FathersIncome, sm.BloodGroup, sm.Address, 
       sm.City, sm.State, sm.DateOfJoining, sm.DateOfRecordCreation, 
       sm.UniqueStudentRegNo, ms.Class, ms.Section, ms.TestTerm, 
       ms.Tamil, ms.English, ms.Maths, ms.Science, ms.SocialScience, 
       ms.Total, ms.Average, ms.Grade
FROM StudentsMaster sm
JOIN StudentsMarksheet ms ON sm.UniqueStudentRegNo = ms.UniqueStudentRegNo
WHERE ms.Average > 80 AND ms.Grade = 'A';
'''
df_1 = pd.read_sql(query_1, conn)

df_1
```

```python
# SQL query to fetch students from low-income families scoring above 75%
query_2 = '''
SELECT sm.StudentName, sm.Gender, sm.DOB, sm.FatherName, sm.MotherName, 
       sm.FathersOccupation, sm.FathersIncome, sm.BloodGroup, sm.Address, 
       sm.City, sm.State, sm.DateOfJoining, sm.DateOfRecordCreation, 
       sm.UniqueStudentRegNo, ms.Class, ms.Section, ms.TestTerm, 
       ms.Tamil, ms.English, ms.Maths, ms.Science, ms.SocialScience, 
       ms.Total, ms.Average, ms.Grade
FROM StudentsMaster sm
JOIN StudentsMarksheet ms ON sm.UniqueStudentRegNo = ms.UniqueStudentRegNo
WHERE sm.FathersIncome < 200000 AND ms.Average > 75;
'''
df_2 = pd.read_sql(query_2, conn)

df_2
```

```python
# SQL query to count students admitted in 2024
query_3 = '''
SELECT COUNT(*) AS StudentCount
FROM StudentsMaster
WHERE strftime('%Y', DateOfJoining) = '2024';
'''
df_3 = pd.read_sql(query_3, conn)
df_3
```
